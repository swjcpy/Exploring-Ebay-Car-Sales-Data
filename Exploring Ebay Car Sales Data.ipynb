We can find the dataset contains 20 columns, most of which are strings. some columns have null values, but none have more than ~20% null values and the column names use camelcase instead of Python's preferred snakecase. Hence we need to convert the column names from camelcase to snakecase and reword some of the column names based on the data dictionary to be more descriptive.

```python
maps = {'yearOfRegistration':'registration_year',
       'monthOfRegistration':'registration_month',
       'notRepairedDamage':'unrepaired_damage',
       'dateCreated':'ad_created',
       'dateCrawled':'date_crawled',
       'offerType':'offer_type',
       'vehicleType':'vehicle_type',
       'powerPS':'power_p_s',
       'fuelType':'fuel_type',
       'nrOfPictures':'nr_of_pictures',
       'postalCode':'postal_code',
       'lastSeen':'last_seen'}
autos = autos.rename(columns=maps)
autos.columns
```

```python
autos.describe(include='all')
```

From above, we can find that the price and odometer columns are numeric values stored as text. Hence we need to convert them to int datatype and rename the column to odometer_km

```python
autos['price']=autos['price'].str.replace('$','').str.replace(',','').astype(int)
```

```python
autos['odometer']=autos['odometer'].str.replace('km','').str.replace(',','').astype(int)
```

```python
autos[:10]
```

```python
autos = autos.rename(columns={'odometer':'odometer_km'})
```

```python
brands = autos['brand'].unique()
```

```python
brand={}
for i in brands:
    isthebrand = autos['brand'] == i
    mean_price = autos[isthebrand]['price'].mean()
    brand[i] = mean_price
brand
```

```python
mileage={}
for i in brands:
    isthebrand = autos['brand'] == i
    mean_price = autos[isthebrand]['odometer_km'].mean()
    mileage[i] = mean_price
mileage
```

```python
brand_series = pd.Series(brand)
mileage_series = pd.Series(mileage)
price_df = pd.DataFrame(brand_series,columns=['mean_price'])
mileage_df = pd.DataFrame(mileage_series,columns=['mean_mileage'])
```
